```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from sklearn.cluster import KMeans
```

```python
img = cv2.imread("strange.jpg")

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
points = img.reshape(-1,3)
```

```python
points.shape
```

```python
model = KMeans(4)
```

```python
model.fit(points)
```

```python
pallets = model.cluster_centers_.astype(int)
```

```python
plt.imshow(pallets.reshape(1, -1, 3))
```

```python
blank = np.zeros(img.shape, dtype=int)
```

```python
colors = model.labels_.reshape(img.shape[:2])
```

```python
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        
        blank[i, j] = pallets[colors[i, j]]
```

```python
plt.imshow(blank)
```
